```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='darkgrid')
f, axes = plt.subplots(2,2, figsize=(7,7), sharex=True)
sns.despine(left=True)

# Plot a simple histogram with binsize determined automatically
sns.distplot(nba.FG_PCT, kde=False, color="b", ax=axes[0,0])
# Plot a kernel density estimate and rug plot
sns.distplot(nba.FG_PCT, hist=False, rug=True, color="r", ax=axes[1,0])
# Plot a filled kernel density estimate
sns.distplot(nba.FG_PCT, hist=False, color="g", ax=axes[0,1])
# Plot a historgram and kernel density estimate
sns.distplot(nba.FG_PCT, color="m", ax=axes[1,1])

plt.setp(axes, yticks=[])
plt.tight_layout()
plt.show()
```

```python
## A scatterplot showing the relationship between a players FG_PCT and MIN (minutes played)

## A scatterplot is useful in determining whether there may or may not be a relationship between two
## variables. In this plot, we see there may be a correlation between minutes played and field goal percentage,
## suggesting that as minutes played increases, the average player's field goal percentage gets closer to 0.5.

## Not much can be said in terms of distribution and you would need to use additional tools to determine if there is
## in fact a relationship between the two variables.
```

```python
# Define a palette to ensure that colors will be
# shared across the facets
palette = dict(zip(nba.TEAM_ABBREVIATION.unique(),
                   sns.color_palette("rocket_r", 30)))

# Plot the lines on two facets
sns.relplot(x="MIN", y="SALARY_MILLIONS",
            size="AGE", hue="TEAM_ABBREVIATION",
            size_order=["T1", "T2"], palette=palette,
            height=5, aspect=.75, facet_kws=dict(sharex=False),
            kind="line", legend="full", data=nba)
plt.show()
```

```python
sns.violinplot(x="TEAM_ABBREVIATION", y="MIN", data=nba[nba.TEAM_ABBREVIATION.isin(['OKC','LAL','PHI'])])
plt.show()
```

```python
sns.stripplot(x="TEAM_ABBREVIATION", y="MIN", data=nba[nba.TEAM_ABBREVIATION.isin(['OKC','LAL','PHI'])])
plt.show()
```

```python
sns.catplot(x="TEAM_ABBREVIATION", y="MIN", data=nba[nba.TEAM_ABBREVIATION.isin(['OKC','LAL','PHI'])]);
plt.show()
```

```python
sns.boxplot(x="TEAM_ABBREVIATION", y="MIN", data=nba[nba.TEAM_ABBREVIATION.isin(['OKC','LAL','PHI'])])
plt.show()
```
